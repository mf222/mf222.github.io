Jupyter para obtener intervalos de tiempo de los puntos para poder generar un heatmap 

```python
import pandas as pd
import numpy as np
import json


points = pd.read_csv("../data/device_positions.csv")
points['DATE1'] = pd.to_datetime(points['DATE1'])

points['DOW'] = points['DATE1'].dt.dayofweek

dow = {0: 'Lunes', 1:'Martes', 2:'Miercoles', 3:'Jueves', 4:'Viernes', 5:'Sabado', 6:'Domingo'}
file_name = 'data_'

fieldnames = ("INDEX","ID","LAT","LONG", "DATE1", "DOW")

# print(points.loc[points['DOW'] == 1])
del points['DATE2']   

def write_csv_dow():
    for i in range(7):
        rows = points.loc[points['DOW'] == i]
        rows.to_csv('../data/' + file_name + '{}.csv'.format(dow[i]))
    print('Termine')

write_csv_dow()

points['count'] = np.zeros_like(points['INDEX'])
points_agg_all = points.groupby(['INDEX', pd.Grouper(key='DATE1', freq='10min')]).agg({'count': lambda x: x.count()})

points_agg_all['LAT'] = points['LAT']

points_agg_all
# print(points)

# print(points['serie'])

```
